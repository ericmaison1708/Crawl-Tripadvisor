# Lấy mã FIPS cho các thành phố

```python
import pandas as pd
```

```python
df = pd.read_csv('fips_code.txt', delimiter='\t', header=None)
df
```

```python
# Xóa khoảng trắng dư thừa trong tất cả các cột
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Sau đó, tách cột FIPS và City
df[['FIPS', 'City']] = df[0].str.split(r'\s+', n=1, expand=True)

# Tiếp theo, xử lý cột FIPS
df['FIPS'] = df['FIPS'].str.extract(r'([A-Z]{2}\d{1}[A-Z]?\d?)')
df
```

```python
df['FIPS'] = df[0].str[:4]
df = df[~df['FIPS'].str.isnumeric()]
```

```python
df.drop_duplicates().dropna()
```

```python
df = df.drop(columns=[0])
df
```

```python
# Loại bỏ các dòng có giá trị NaN trong cột FIPS
df = df.dropna(subset=['FIPS'])
df
```

```python
# Loại bỏ tất cả các số trong cột City
df['City'] = df['City'].str.replace(r'\d+', '', regex=True)

# Loại bỏ khoảng trắng thừa
df['City'] = df['City'].str.strip()

df
```

```python
# Loại bỏ "City of", dấu ngoặc kép và dấu phẩy
df['City'] = df['City'].str.replace(r'\s*City of\s*', '', regex=True)  # Xóa "City of"
df['City'] = df['City'].str.replace(r'[",]+', '', regex=True)  # Xóa dấu ngoặc kép, khoảng trắng và dấu phẩy
df['City'] = df['City'].str.strip()  # Xóa khoảng trắng thừa

# Đưa tên thành phố về dạng viết hoa chuẩn
df['City'] = df['City'].str.title()
```

```python
df.to_csv("fips_code.csv", index=False)
```
